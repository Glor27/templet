```c++
#pragma cling add_include_path("../../lib/")

const int DELAY = 1.0;
const int NUM_WORKERS = 20;
const int NUM_TASKS = 20;

#include "blchsym.hpp"

using namespace std;
{
    exec_order::define();
    exec_order::print();
    
    std::cout << std::endl;   
    
 	engine eng;
	basesim_engine teng;

    bchain bch(eng);
    stopper stop(eng);
    bworker w[NUM_WORKERS];
    
    for(int i=0; i<NUM_WORKERS; i++){
        w[i].engines(eng,teng);
        w[i].worker_ID = i;
        bch.in(w[i].out);
        stop.in(w[i].ready);
    }
    
    stop.num_total = NUM_WORKERS;

	eng.start();
	teng.run();

	if (eng.stopped()) {
		std::cout << std::endl;        
		std::cout << "Success!!!" << std::endl;
		
		std::cout << "Maximum number of tasks executed in parallel : " << teng.Pmax() << std::endl;
		std::cout << "Time of sequential execution of all tasks    : " << NUM_TASKS * DELAY << std::endl;
		std::cout << "Time of parallel   execution of all tasks    : " << teng.Tp() << std::endl;
        std::cout << "Overworking due to reduntant  execution      : " << teng.T1() / NUM_TASKS << std::endl;
		std::cout << "Speedup without com/sync drawbacks           : " << NUM_TASKS * DELAY / teng.Tp() << std::endl;
		std::cout << "Efficiency without com/sync drawbacks        : " << NUM_TASKS * DELAY / teng.Tp() / NUM_WORKERS << std::endl;
	}
	else std::cout << "Failure..." << std::endl;
}
```
