```c++
#pragma cling add_include_path("../../lib/")
#include "blchsym.hpp"

using namespace std;
{
    exec_order::define();
    exec_order::print();
    
    std::cout << std::endl;   
    
 	engine eng;
	basesim_engine teng;

    bchain bch(eng);
    stopper stop(eng);
    bworker w[NUM_WORKERS];
    
    for(int i=0; i<NUM_WORKERS; i++){
        w[i].engines(eng,teng);
        w[i].worker_ID = i;
        bch.in(w[i].out);
        stop.in(w[i].ready);
    }
    
    stop.num_total = NUM_WORKERS;

	eng.start();
	teng.run();

	if (eng.stopped()) {
		std::cout << std::endl;        
		std::cout << "Success!!!" << std::endl;
		
		std::cout << "Maximum number of tasks executed in parallel : " << teng.Pmax() << std::endl;
		std::cout << "Time of sequential execution of all tasks    : " << teng.T1() << std::endl;
		std::cout << "Time of parallel   execution of all tasks    : " << teng.Tp() << std::endl;
		std::cout << "Speedup without drawbacks (max possible)     : " << teng.T1()/teng.Tp() << std::endl;
		std::cout << "Efficiency without drawbacks (max possible)  : " << teng.T1()/teng.Tp()/NUM_WORKERS << std::endl;
	}
	else std::cout << "Failure..." << std::endl;
}
```
