```c++
#pragma cling add_include_path("../../lib/")

const int N = 3;

double A[N][N] = { {1.0,2.0,3.0},
                   {1.0,2.0,3.0},
                   {1.0,2.0,3.0}};

double B[N][N] = { {1.0,1.0,1.0},
                   {2.0,2.0,2.0},
                   {3.0,3.0,3.0}};

double C[N][N] = { 0.0 };

#include "3-1-ringmult.hpp"

{
	engine eng;
    
    Stopper stopper(eng);
    Link l[N];
        
    for(int i=0; i<N; i++){
        l[i].engines(eng); stopper.in(l[i].stop);        
        l[i].next = &l[(i+1)%N]; 
        
        l[i].i = i; l[i].out.j = i; 
    }

	eng.start();
    
    for(int i=0; i<N; i++){
        for(int j=0; j<N; j++){
            cout << "  C[" << i << "," << j << "] = " << C[i][j];
        }
        cout << endl;
    }

	if (eng.stopped()) std::cout << "Success!!!" << std::endl;
    else std::cout << "Failure..." << std::endl;
}
```
