```c++
#pragma cling add_include_path("../../lib/")

const int N = 10;
int arr[N];

#include "4-1-pairwiseNNS.hpp"

struct helper{
    helper(engine&e):eng(&e){}  
    engine*eng;
    block* block_arr[N];
    
    void start(){
        for(int i=0; i<N; i++){
            Start*s = new Start(*eng);
            s->out.blockID = i;
            block_arr[i] = &(s->out);
        }
    }
    
    void pair(int n1, int n2){
        Pair* p = new Pair(*eng);
        p->in1(*block_arr[n1]); p->in2(*block_arr[n2]);
        block_arr[n1] = &(p->out1); block_arr[n2] = &(p->out2);
    }
    
    void stop(){
        Stop* s = new Stop(*eng);
        for(int i=0; i<N; i++) s->in(*block_arr[i]);
    }
};

engine eng; 
helper help(eng);

srand(time(NULL));
for(int i=0; i<N; i++){ arr[i] = rand()%N; cout << arr[i] << " ";} cout << endl;        

help.start();

for(int b=0; b < N; b++){
    for(int i=0; i < b; i++){ help.pair(i,b); cout << "[" << i+1 <<","<< b+1 << "] ";}
    cout << endl;
}

cout << endl;

help.stop();

eng.start();

for(int i=0; i<N; i++) cout << arr[i] << " ";  cout << endl;

if (eng.stopped()) std::cout << "Success!!!" << std::endl;
    else std::cout << "Failure..." << std::endl;
```
