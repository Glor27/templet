```c++
#include <stdlib.h>
#include <iostream>
#include <fstream>
#include <string>
using namespace std;
{
    system("g++ pythagor.cpp -I ~/lib -o pythagor 2> out.txt");
    ifstream file("out.txt"); bool ok=true; for (string line; getline(file, line); ){cout << line << endl; ok = false;}
    if(ok)cout << "Compilation is Ok";
    system("rm out.txt");
}
```
